```python
import textacy
from spacy import displacy
```

```python
text = '''Elon Musk, the visionary entrepreneur and CEO of SpaceX, has always had a passion for space exploration. 
Born in Pretoria, South Africa, Musk grew up with dreams of reaching the stars. After making a name for himself in 
the tech industry with companies like Zip2 and PayPal, Musk set his sights on revolutionizing the space industry.'''
```

```python
nlp = textacy.load_spacy_lang('en_core_web_sm')
doc = textacy.make_spacy_doc(text, nlp)
doc._.preview
```

```python
displacy.render(doc)
```

```python
displacy.render(doc, style='ent')
```

```python
displacy.render(doc, style='span')
```
