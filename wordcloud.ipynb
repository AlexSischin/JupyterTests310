```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
with open('./data/WarAndPeace/content.txt') as file:
    text = file.read()
text[10000:11000]
```

```python
wc = WordCloud(max_font_size=40, mode='RGBA', background_color=None).generate(text)
plt.imshow(wc)
plt.axis("off")
```
